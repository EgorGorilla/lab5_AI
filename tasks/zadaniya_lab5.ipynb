# **4. Добавление новых столбцов и строк**

### Используя Таблицу 1:
### 4.1. Добавьте новый столбец "Категория зарплаты":
### • "Низкая" - если Зарплата $<$ 60000.
### • "Средняя" - если 60000 $\leq$ Зарплата $<$ 100000.
### • "Высокая" - если Зарплата $\geq$ 100000.

```python
df_csv["Категория зарплаты"] = df_csv["Зарплата"].apply(
    lambda x: "Низкая" if x < 60000 
              else "Средняя" if 60000 <= x < 100000 
              else "Высокая"
)
display(df_csv)
```

### Метод isna используется для определения, является ли значение пропущенным. Он возвращает True для ячеек с пустыми значениями (NaN) и False для всех остальных.
